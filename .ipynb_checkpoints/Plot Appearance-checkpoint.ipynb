```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#ax.plot(x,y,color='black')
#ax.plot(x,y,color='#FF8C00',linewidth=10,alpha=10,linestyle='--',marker = '+',markersize=15
#        markerfacecolor='yellow',markeredgewidth=5,markeredgecolor='red')   #RGB Hex Codes to google search
ax.plot(x,y,color = 'blue', lw = 2, alpha=5,ls='-',marker='o',ms=15,mfc='yellow',mew=3,mec='red')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,y,color='red', lw=3, ls='--')
ax1.set_xlim([0,2])   #[Lower Bound, Upper bound]
ax1.set_ylim([0,5])
```

```python
fig.savefig('sridhar.jpg',dpi=300)
```
