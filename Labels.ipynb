```python
plt.show()
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'g--')
#plt.show(x,y)
plt.xlabel('Numbers')
plt.ylabel('Squares')
plt.title('Plot')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
```

```python
plt.subplot(1,2,1)
plt.plot(y,x)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.9,0.9])    #[Left,Bottom,width and Height]
axes.plot(x,y)
axes.set_xlabel('X-axes')
axes.set_ylabel('Y-axes')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.15,0.4,0.4,0.4])

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(x,y)
axes2.set_title('Smaller Plot')
```

```python
fig,axes = plt.subplots(nrows = 1,ncols=2)
#plt.tight_layout()               #To Overcome overlaps
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
#for current_ax in axes:
 #   current_ax.plot(x,y)
```
